````
AHS-3: PRIMCAREPROV_V5_2016.xlsx, primcareprov_v5_2016.sasbdat
Persons with a usual primary care provider (Composite variable created based on the following variables)
```
```
ACCELI42 - PERS ELIGIBLE FOR ACCESS SUPPLEMENT-R4/2 (-1 INAPPLICABLE, 1 YES)
HAVEUS42 - AC05 DOES PERSON HAVE USC PROVIDER-R4/2 (-9,-8,-7,-1, 1 YES, 2 NO)
LOCATN42 - USC LOCATION-R4/2 (-9,-8,-7,-1, 1 OFFICE, 2 HOSPITAL NOT ER, 3 HOSPITAL ER)
MINORP42 - AC22 GO TO USC FOR NEW HEALTH PROB-R4/2 (-8,-7,-1, 1 YES, 2 NO)
PREVEN42 - AC22 GO TO USC FOR PRVNTVE HLT CARE-R4/2 (-8,-7,-1, 1 YES, 2 NO)
REFFRL42 - AC22 GO TO USC FOR REFERRALS-R4/2 (-8,-7,-1, 1 YES, 2 NO)
```

`````python
import saspy
`````

`````python
%%SAS
LIBNAME pufmeps  'S:\CFACT\Shared\PUF SAS files\SAS V8' access=readonly;
options nocenter ls=132;
%let vars_kept = ACCELI42 HAVEUS42 LOCATN42 MINORP42 PREVEN42 REFFRL42 saqwt16f varstr varpsu; 
Data h192;
  set pufmeps.h192 (keep = &vars_kept);
  if ACCELI42=1 and haveus42=1 and locatn42 in (1,2) then do;
    if MINORP42=1 and PREVEN42=1 and REFFRL42=1 then primcareprov=1;
    else if MINORP42=2 or PREVEN42=2 or REFFRL42=2 then primcareprov=0;
    else primcareprov=.;
  end;
  else if (haveus42=1 and locatn42=3) or haveus42=2 then primcareprov=0;
run;
proc freq data=h192;
   tables primcareprov /list nopercent nocum;
run;
ods listing close;
ods graphics off;
ods exclude summary;
proc surveymeans data=h192 nobs  sumwgt mean stderr sum ;
  stratum varstr;
  cluster varpsu;
  weight saqwt16f;
  var  primcareprov;
run;
`````

`````python
%%SAS
LIBNAME pufmeps  'S:\CFACT\Shared\PUF SAS files\SAS V8' access=readonly;
options nocenter ls=132;
%let vars_kept = ACCELI42 HAVEUS42 LOCATN42 MINORP42 PREVEN42 REFFRL42 saqwt16f varstr varpsu; 
Data h192;
  set pufmeps.h192 (keep = &vars_kept);
  if ACCELI42=1 and haveus42=1 and locatn42 in (1,2) then do;
    if MINORP42=1 and PREVEN42=1 and REFFRL42=1 then primcareprov=1;
    else if MINORP42=2 or PREVEN42=2 or REFFRL42=2 then primcareprov=0;
    else primcareprov=.;
  end;
  else if (haveus42=1 and locatn42=3) or haveus42=2 then primcareprov=0;
  if primcareprov in (0,1) then subpop=1;
run;
proc freq data=h192;
   tables primcareprov subpop/list nopercent nocum;
    where saqwt16f >0;
run;
ods listing close;
ods graphics off;
ods exclude summary;
proc surveymeans data=h192 nobs  sumwgt mean stderr sum ;
  stratum varstr;
  cluster varpsu;
  weight saqwt16f;
  var  primcareprov;
domain subpop('1');
run;
`````

`````python
%%SAS
LIBNAME pufmeps  'S:\CFACT\Shared\PUF SAS files\SAS V8' access=readonly;
options nocenter ls=132;
%let vars_kept = MDUNAB42 MDDLAY42 DNUNAB42 DNDLAY42 PMUNAB42 PMDLAY42 saqwt16f varstr varpsu; 
Data h192;
  set pufmeps.h192 (keep = &vars_kept);
    array x[3] MDUNAB42 DNUNAB42 PMUNAB42;
    array y[3] MDDLAY42 DNDLAY42 PMDLAY42;
    array z[3] MEDDIFF DENTDIFF PMEDDIFF;
    
    do i = 1 to 3;
      if x[i] = 1 or y[i] = 1 then z[i] =1;
      else if x[i] = 2 and y[i] = 2 then z[i] =0;
    end;
  
  run;
proc freq data=h192;
   tables MEDDIFF DENTDIFF PMEDDIFF /list missing nopercent nocum;
    where saqwt16f >0;
run;
ods listing close;
ods graphics off;
ods exclude summary;
proc surveymeans data=h192 nobs  sumwgt mean stderr sum ;
  stratum varstr;
  cluster varpsu;
  weight saqwt16f;
  var  MEDDIFF DENTDIFF PMEDDIFF ;
*domain subpop('1');
run;
`````
